Don't write your own shuffle and split functions; use sklearn's classes

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# loads data
train_pd = pd.read_csv("../data/raw/train.csv.zip", compression="zip")
test_pd = pd.read_csv("../data/raw/test.csv.zip", compression="zip")
```

```python
# test basic train test split
from sklearn.model_selection import train_test_split

train, dev = train_test_split(train_pd, test_size = 0.5)

# print(train.shape)
# print(type(train))
# print(train.head(3))
train.Category.value_counts().plot.bar()
plt.show()

# print(dev.shape)
# print(type(dev))
# print(dev.head(3))
dev.Category.value_counts().plot.bar()
plt.show()
```

```python
# test train test split with stratification, to guarantee ratios of Categories
from sklearn.model_selection import train_test_split

train, dev = train_test_split(train_pd, test_size = 0.5, stratify = train_pd.Category)

# print(train.shape)
# print(type(train))
# print(train.head(3))
train.Category.value_counts().plot.bar()
plt.show()

# print(dev.shape)
# print(type(dev))
# print(dev.head(3))
dev.Category.value_counts().plot.bar()
plt.show()
```

```python
# test train test split, but only shuffle, no split
from sklearn.model_selection import train_test_split

train, dev = train_test_split(train_pd, test_size = 0, shuffle = True, random_state = 0)

print(train_pd.head(3))

print(train.shape)
print(train.head(3))

print(dev.shape)
```
