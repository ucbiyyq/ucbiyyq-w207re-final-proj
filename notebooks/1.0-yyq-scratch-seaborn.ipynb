```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
```

```python
# loads data
train_pd = pd.read_csv("../data/raw/train.csv.zip", compression="zip")
test_pd = pd.read_csv("../data/raw/test.csv.zip", compression="zip")
```

```python
dat = train_pd.Category.value_counts().plot.bar()
```

```python
dat = train_pd.groupby(["Category"]).size().reset_index(name='counts').sort_values(by = "counts", ascending = False)

g = sns.barplot(x="Category", y="counts", data=dat)
_ = g.set_xticklabels(g.get_xticklabels(), rotation = 90)
```

```python
# https://www.tutorialspoint.com/seaborn/seaborn_facet_grid.htm
cats = train_pd.Category.unique()
city = train_pd[["Category", "X", "Y"]].query("Y < 50")

g = sns.FacetGrid(city, col = "Category", col_wrap = 4, height = 3, sharex = True, sharey = False)
g.map(sns.distplot, "X", kde = False)
plt.show()
```

```python
g = sns.FacetGrid(city, col = "Category", col_wrap = 4, height = 3)
g.map(sns.distplot, "X", kde = False)
plt.show()
```
