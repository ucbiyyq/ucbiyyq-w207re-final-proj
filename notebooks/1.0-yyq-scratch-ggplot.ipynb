```python
import numpy as np
import pandas as pd
from ggplot import *
```

```python
# loads data
train_pd = pd.read_csv("../data/raw/train.csv.zip", compression="zip")
test_pd = pd.read_csv("../data/raw/test.csv.zip", compression="zip")
```

```python
dat = train_pd.groupby(["Category"]).size().reset_index(name='counts')

ggplot(aes(x = "Category", y = "counts"), data = dat) + \
    geom_bar()
```

```python
dat = train_pd[["Category", "X", "Y"]].query("Y < 50")

ggplot(aes(x = "X", y = "Y"), data = dat) + \
    geom_point(alpha = 0.1) + \
    facet_wrap("Category") + \
    theme(plot_margin = dict(right = 2, top = 2))
```
